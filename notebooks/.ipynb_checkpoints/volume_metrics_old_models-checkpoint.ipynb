In this notebook we analyze the structural quality of 3D defect segmentation models using a set of volumetric metrics computed from the reconstructed log volume.

| Metric | Meaning |
|--------|---------|
| **Volume (cm³)** | Total estimated defect volume. Detects under/over-segmentation. |
| **Components** | Number of connected 3D components. Detects fragmentation or merging. |
| **Continuity** | Ratio of largest connected component to total volume. Measures structural smoothness. |
| **Compactness** | Shape irregularity measure derived from 3D surface complexity. |

```python
import sys, os
sys.path.append(os.path.abspath("../"))

import numpy as np
import pandas as pd
from tqdm import tqdm
import seaborn as sns
import matplotlib.pyplot as plt

from src.pipelines.segmentation_pipeline import SegmentationPipeline
from src.visualization.volume_metrics import compute_volume_metrics, CLASS_LABELS
```

```python
BASE = os.path.abspath(os.path.join(os.path.abspath("../")))

LOGS = os.path.join(BASE, "logs")

EVAL_MODES = {
    "OLD_MODEL": {
        "ckpt": os.path.join(LOGS, "old_model", "checkpoints", "best.pt"),
        "use_3d": False,
    },
    "OLD_MODEL_3DMRF": {
        "ckpt": os.path.join(LOGS, "old_model", "checkpoints", "best.pt"),
        "use_3d": True,
    },
    "OLD_MODEL_MRF2D": {
        "ckpt": os.path.join(LOGS, "old_model_mrf", "checkpoints", "best.pt"),
        "use_3d": False,
    },
    "OLD_MODEL_MRF2D_3DMRF": {
        "ckpt": os.path.join(LOGS, "old_model_mrf", "checkpoints", "best.pt"),
        "use_3d": True,
    },
}
```

```python
TEST_TIFFS = sorted([
    os.path.join("../data/volume_metric_test", f)
    for f in os.listdir("../data/volume_metric_test")
    if f.lower().endswith((".tiff", ".tif"))
])

len(TEST_TIFFS)
```

```python
def evaluate_model_on_tiff(model_path, tiff_path, use_3d=False):
    pipeline = SegmentationPipeline(
        model_type="cnn",
        use_mrf=use_3d,  # this toggles ONLY 3D MRF
    )

    refined_volume, _, metrics, anomalies = pipeline.run(
        tiff_path=tiff_path,
        model_path=model_path,
        visualize=False,
        return_metrics=True,
    )

    return metrics, anomalies
```

For each defect class we compute dataset-level statistics:

- mean volume ± std  
- mean component count ± std  
- mean continuity ± std  
- mean compactness ± std  

These values define expected behavior for each defect type.

During evaluation, each segmentation is compared against these expected distributions:

- **Values outside of standard deviations are flagged as anomalies**
- **Each class contributes to an anomaly count**

Lower anomaly count = more stable and physically plausible 3D reconstruction.

```python
rows = []

for mode_name, cfg in EVAL_MODES.items():
    print(f"\nEvaluating {mode_name}...\n")

    for tiff_path in tqdm(TEST_TIFFS):
        metrics, anomalies = evaluate_model_on_tiff(
            model_path=cfg["ckpt"],
            tiff_path=tiff_path,
            use_3d=cfg["use_3d"],   # ONLY 3D MRF is controlled here
        )

        for cls in CLASS_LABELS.keys():
            m = metrics[cls]
            an = anomalies[cls]

            rows.append({
                "model": mode_name,
                "tiff": os.path.basename(tiff_path),
                "class": CLASS_LABELS[cls],

                "volume": m["volume_cm3"],
                "components": m["components"],
                "continuity": m["continuity"],
                "compactness": m["compactness"],

                "anomaly_count": len(an),
                "anomaly_text": "; ".join(an),
            })

df = pd.DataFrame(rows)
```

```python
pd.set_option("display.float_format", lambda x: f"{x:.2f}")

model_summary = df.groupby("model")[
    ["volume", "components", "continuity", "compactness", "anomaly_count"]
].mean()

model_summary
```

```python
class_summary = df.groupby(["model", "class"])[
    ["volume", "components", "continuity", "compactness", "anomaly_count"]
].mean()

class_summary
```

```python
anomaly_rate = df.groupby("model")["anomaly_count"].mean().sort_values()
anomaly_rate
```

```python
pivot = df.pivot_table(
    index="class", columns="model", values="volume", aggfunc="mean"
)

plt.figure(figsize=(10,5))
sns.heatmap(pivot, annot=True, fmt=".1f", cmap="viridis")
plt.title("Average Defect Volume per Model")
plt.show()
```
