```python
import sys, os
sys.path.append(os.path.abspath("../"))

import torch
import matplotlib.pyplot as plt

from src.train import train_model
from src.evaluate import evaluate_model, visualize_slices
from src.datamodule import make_dataloaders
```

```python
model, (train_losses, val_losses, val_ious) = train_model(
    root_dir="../data/old_dataset",
    out_dir="../logs/old_model_mrf",
    epochs=200,
    batch_size=8,
    lr=3e-4,
    size=256,
    num_classes=7,
    early_stop_patience=15,
    use_mrf=True,
    use_crf=False,
    use_augmentation=True,
)

plt.plot(train_losses, label="Train Loss")
plt.plot(val_losses, label="Val Loss")
plt.plot(val_ious, label="Val IoU")
plt.legend()
plt.show()
```

```python
weights_path = "../logs/old_model_mrf/checkpoints/best.pt"
root_dir = "../data/old_dataset"

mean_iou, per_class_ious = evaluate_model(
    weights_path=weights_path,
    root_dir=root_dir,
    num_classes=7,
    batch_size=2,
    size=256,
    use_mrf=True,
    use_crf=False,
)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
_, val_loader = make_dataloaders(
    root_dir=root_dir,
    batch_size=2,
    size=(256, 256),
    use_augmentation=True,
)
visualize_slices(
    model.to(device),
    val_loader,
    device=device,
    n_slices=6,
    start_index=0,
    use_mrf=True,
    use_crf=False,
)
```
