```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Class": ["Background", "Obvod", "Hniloba", "Dutina", "HrcaZ", "HrcaN", "Trhlina", "Mean IoU"],
    "OldModel Raw": [0.9833, 0.9752, 0.9253, 0.8931, 0.5963, 0.6894, 0.6188, 0.8116],
    "OldModel CRF":          [0.9836, 0.9705, 0.8860, 0.8930, 0.5130, 0.6579, 0.6041, 0.7869],
    "OldModel 2DMRF":        [0.9838, 0.9745, 0.9241, 0.8993, 0.5450, 0.7355, 0.6780, 0.8200]
}

df = pd.DataFrame(data)
df.set_index("Class", inplace=True)


def highlight_changes(row):
    base = row["OldModel Raw"]
    return [
        "background-color: lightgreen" if val > base else
        "background-color: salmon" if val < base else ""
        for val in row
    ]


styled_df = (
    df.style
      .apply(highlight_changes, axis=1)
      .format("{:.4f}")
)

styled_df
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, fmt=".4f")
plt.title("Per-class IoUs across models")
plt.tight_layout()
plt.show()
```
